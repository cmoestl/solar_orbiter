```
path = '/content/drive/My Drive/solo/mag/LL02/2020/'
files = os.listdir(path)
files.sort()
files = [os.path.join(path, f) for f in files]
print(files[4])

#https://github.com/MAVENSDC/cdflib
f1 = cdflib.CDF(files[4])



path2 = '/content/drive/My Drive/solo/swa/LL02/2020/'
files2 = os.listdir(path2)
files2.sort()
files2 = [os.path.join(path, f) for f in files2]
print(files2[0])

p1 = cdflib.CDF(files2[0])
print(p1.cdf_info())
```

```
f1.cdf_info()
```

```
b=f1.varget('B_RTN')
br=b[:,0]
bt=b[:,1]
bn=b[:,2]
btot=np.sqrt(br**2+bt**2+bn**2)

#print(f1.attget(attribute='UNITS',entry='EPOCH'))

time=f1.varget('EPOCH')
print(time[0], time[1])

t1=parse_time(cdflib.cdfastropy.convert_to_astropy(time, format=None)).datetime


t1
```
